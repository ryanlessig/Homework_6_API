## Generate Cities List

```python
#import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

#import API key
from api_keys import api_key

#import citipy  wrapper to determine city names based on coordinates
from citipy import citipy

#setup the range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
#create lists for holding the latitude and longitude of cities and their names
lat_lngs = []
cities = []

#generate a random set of latitude and longitude combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)

#find the nearest city for each latitude and longitude combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    #if the found city name is not already in the cities list, then add the city name to the cities list
    if city not in cities:
        cities.append(city)

#print the length of the cities list to ensure we have enough city names
len(cities)
```

### Perform API Calls

```python
#assign base URL
url = "http://api.openweathermap.org/data/2.5/weather?"

#update the base URL with the API key and start the query string
query_url = url + "units=imperial" + "&appid=" + api_key  + "&q="

#create lists to store values from the API call
City = []
Cloudiness = []
Country = []
Date = []
Humidity = []
Latitude = []
Longitude = []
Max_Temperature = []
Wind_Speed = []

#initialize the number of cities at 1
city_counter = 1

#print header of data retrieval process
print(f'Beginning Data Retrieval')
print(f'-----------------------------')

#loop through each city in the cities list, perform an API call to find the city and it's corresponding weather info.
#if the city is not found, go to the except line and print out a statement saying the city was not found.
for city in cities:
    try:
        api_data = requests.get(query_url + city).json()
        City.append(api_data['name'])
        Cloudiness.append(api_data['clouds']['all'])
        Country.append(api_data['sys']['country'])
        Date.append(api_data['dt'])
        Humidity.append(api_data['main']['humidity'])
        Latitude.append(api_data['coord']['lat'])
        Longitude.append(api_data['coord']['lon'])
        Max_Temperature.append(api_data['main']['temp_max'])
        Wind_Speed.append(api_data['wind']['speed'])
        
        #print out a processing line to tell the user which city the API call is trying to find
        print(f'Finding city #{city_counter} of {len(cities)} | City Name: {city}')
        
        #once the city is found, increment the city counter by 1
        city_counter += 1
    
    #if the city name is not found in the API call, print out a statement to tell the user.
    #increment the city counter by 1
    except:
        print(f'There was no city by the name of {city}. Trying the next city in the list.')
        city_counter += 1
    continue

#print footer of data retrieval process
print(f'-----------------------------')
print(f'Data Retrieval Complete')
print(f'-----------------------------')
```

### Convert Raw Data to DataFrame

```python
#put all the data retrieved into a dataframe named city_weather_data
city_weather_data = pd.DataFrame({'City': City,
                                  'Cloudiness': Cloudiness,
                                  'Country': Country,
                                  'Date': Date,
                                  'Humidity': Humidity,
                                  'Lat': Latitude,
                                  'Lng': Longitude,
                                  'Max Temp': Max_Temperature,
                                  'Wind Speed': Wind_Speed})

#print the first 5 rows of the dataframe
city_weather_data.head()
```

```python
#check the counts of each column to ensure we have more than 500 entries and all entries are complete.
city_weather_data.count()
```

```python
#check for any outliers. In one of the data sets, there was a humidity abouve 100%, which cannot happen.
city_weather_data.describe()
```

```python
#remove row of data where humidity is above 100%
city_weather_data = city_weather_data.drop(city_weather_data[city_weather_data.Humidity > 100].index)
```

```python
#check that we still have more than 500 entries
city_weather_data.count()
```

```python
#export the cities data to a CSV file, no index, include the header
city_weather_data.to_csv("../Output/cities.csv", index=False, header=True)
```
